```python
%%sql

CREATE TABLE pacientes(Pregnancies INT, Glucose INT, BloodPressure INT, SkinThickness INT, Insulin INT,
       BMI DECIMAL(8,2), DiabetesPedigreeFunction DECIMAL(8,2), Age INT, Outcome INT)
```

```python
%%sql
SELECT * FROM pacientes
```

```python
%%sql
INSERT INTO pacientes(Pregnancies,	Glucose	,BloodPressure,	SkinThickness,	Insulin	,BMI,	DiabetesPedigreeFunction,	Age,	Outcome
)
SELECT Pregnancies,	Glucose	,BloodPressure,	SkinThickness,	Insulin,	BMI	,DiabetesPedigreeFunction,	Age	,Outcome 

FROM diabetes WHERE Age>50
```

```python
%%sql

SELECT * FROM pacientes
```

```python
%%sql

ALTER TABLE pacientes
ADD Profile VARCHAR(10)
```

```python
%%sql

SELECT * FROM pacientes
```

```python
%%sql

UPDATE pacientes
SET Profile = 'Normal'
WHERE BMI <30
```

```python
%%sql
SELECT * FROM pacientes
```

```python
%%sql
UPDATE pacientes
SET Profile = 'Obese'
WHERE BMI>=30
```

```python
%%sql
SELECT * FROM pacientes
```
